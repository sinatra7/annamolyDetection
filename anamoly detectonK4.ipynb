```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
from scipy.spatial.distance import cdist
```

```python
data = pd.read_csv('timeslape.csv')
```

```python
numeric_data = data[['Suction_Flow', 'Suction_Pressure', 'Discharge_Flow', 'Discharge_Pressure']].dropna()
timestamps = data.loc[numeric_data.index, ['TimeStamp']]
```

```python
valid_indices = numeric_data.dropna().index.intersection(timestamps.dropna().index)
numeric_data = numeric_data.loc[valid_indices]
timestamps = timestamps.loc[valid_indices]
```

```python

def kmeans(data, k, max_iterations=100):
    np.random.seed(42)
    centroids = data.sample(n=k).to_numpy()

    def assign_clusters(data, centroids):
        distances = np.linalg.norm(data.to_numpy()[:, None] - centroids, axis=2)
        return np.argmin(distances, axis=1)

    def update_centroids(data, clusters, k):
        return np.array([data[clusters == i].mean(axis=0) if len(data[clusters == i]) > 0 else data.sample(n=1).to_numpy() for i in range(k)])

    for _ in range(max_iterations):
        clusters = assign_clusters(data, centroids)
        new_centroids = update_centroids(data, clusters, k)
        if np.allclose(centroids, new_centroids):
            break
        centroids = new_centroids

    cluster_counts = np.bincount(clusters)
    return centroids, clusters, cluster_counts
```

```python
k = 4
centroids, clusters, cluster_counts = kmeans(numeric_data, k)
```

```python
anomaly_threshold = 0.05  # Set to 5%
anomalous_clusters = np.where(cluster_counts < (anomaly_threshold * len(numeric_data)))[0]
anomalies_indices = np.where(np.isin(clusters, anomalous_clusters))[0]
```

```python
if len(anomalies_indices) == 0:
    # Use distance from centroid as an additional criterion
    distances_to_centroids = cdist(numeric_data, centroids)
    point_distances = np.array([distances_to_centroids[i, clusters[i]] for i in range(len(clusters))])
    distance_threshold = np.percentile(point_distances, 95)  # E.g., top 5% of distances
    anomalies_indices = np.where(point_distances > distance_threshold)[0]
```

```python
data_with_clusters = numeric_data.copy()
data_with_clusters['Cluster'] = clusters
```

```python
if len(anomalies_indices) == 0:
    print("No anomalies detected with the current criteria.")
else:
    anomalies_with_timestamp = pd.concat([timestamps.iloc[anomalies_indices], numeric_data.iloc[anomalies_indices]], axis=1)
    anomalies_with_timestamp['Cluster'] = 'Anomaly'
```

```python
# Assuming we've already run the kmeans function and have the 'clusters' and 'cluster_counts' variables

# Define the threshold for anomalous clusters (10% of total data points)
anomaly_threshold = 0.1 * len(numeric_data)

# Identify anomalous clusters
anomalous_clusters = np.where(cluster_counts < anomaly_threshold)[0]

# Count the number of data points in anomalous clusters
num_anomalies = np.sum(cluster_counts[anomalous_clusters])

print(f"Number of anomalies detected: {num_anomalies}")

# If you want to get the percentage of anomalies:
percent_anomalies = (num_anomalies / len(numeric_data)) * 100
print(f"Percentage of anomalies: {percent_anomalies:.2f}%")

# To get anomaly counts for different k values, you can wrap this in a function:
def get_anomaly_count(clusters, cluster_counts, total_points):
    anomaly_threshold = 0.1 * total_points
    anomalous_clusters = np.where(cluster_counts < anomaly_threshold)[0]
    return np.sum(cluster_counts[anomalous_clusters])

# Then use it for each k value:
k_values = [2, 3, 4]
anomaly_counts = {}

for k in k_values:
    centroids, clusters, cluster_counts = kmeans(numeric_data, k)
    anomaly_counts[k] = get_anomaly_count(clusters, cluster_counts, len(numeric_data))

# Print results
for k, count in anomaly_counts.items():
    print(f"For k={k}, number of anomalies: {count}")
```
